##Hierarchical Clustering

3 Things to keep in mind when doing Hierarchical Clustering -
*   Distance b/w datapoint one and other
*   Distance b/w cluster one and other
*   Distance b/w cluster one and datapoint.

It is calculated using **Euclidean Distance**



image.png

###Types to Linkages b/w clusters :
1.  Single Linkage - Minimum distance b/w two datapoints.
2.  Complete Linkage - Maximum distance b/w two datapoints.
3.  Avg/Mean Linkage - Average of the distance b/w all datapoints.
4.  Centroid Linkage - Distance b/w the centroid of one cluster to the other

#K-means

Steps in k-means :
1.  assume a k value, for ex k=3. ( The entire the dataset will be divided into 3 clusters at random).
2.  Calc. centroid for all the clusters.
3.  Calc. the distance b/w all centroids and all datapoints.
4.  Move/Re-assign the datapoint to the nearest centroids.
5.  Re-compute centroids.
6.  Repeat steps 3 to 5, until we don't have to move any datapoints.

###Calc. Centroids


centroid for cluster one = [(x1+y1+z1)/3   (x2+y2+z2)/3 ]


image.png

###Figuring out the correct k-value

In order to figure out the k-values we use Elbow graph or Scree Plot.

we need to calc. avd. distance in a cluster for k=1 to k=6 and plot it to a elbow chart and wherever the similarity stops increasing drastically that is our appropriate k-value.

#Clustering Code

```python
#import hierarchical clustering lib.
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
Univ = pd.read_csv("Universities.csv")
```

```python
#Normalization Function
def norm_func(i):
  x= (i-i.min())/(i.max()-i.min())
  return (x)
```

```python
#Normalized data frame (Considering the numerical part of data)
df_norm = norm_func(Univ.iloc[:,1:])
```

```python
#Create dendogram
dendrogram = sch.dendrogram(sch.linkage(df_norm,method = 'single'))
```

```python
#create clusters
hc = AgglomerativeClustering(n_clusters =4,
                             affinity = 'euclidean',
                             linkage = 'single')
```

```python
#save clusters for chart
y_hc = hc.fit_predict(df_norm)
Clusters=pd.DataFrame(y_hc, columns = ["Clusters"])
```

```python
Univ['h_clusterid'] = Clusters
Univ
```

#K-means Code

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
```

```python
Univ = pd.read_csv("Universities.csv")
```

```python
#Normalization function
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_Univ_df = scaler.fit_transform(Univ.iloc[:,1:])
```

```python
#How to find optimum no. of Clusters
#The K-means algorithm aims to choose centroids that minimise the inertia, or within cluster sum-of-square
```

```python
wcss = []
for i in range (1,11):
  kmeans = KMeans(n_clusters =i,random_state =0)
  kmeans.fit(scaled_Univ_df)
  wcss.append(kmeans.inertia_)

plt.plot(range(1,11),wcss)
plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
#Optimal value for is b/w 4 to 6
```

```python
#Build Cluster algorithm
from sklearn.cluster import KMeans
clusters_new = KMeans(4, random_state=42)
clusters_new.fit(scaled_Univ_df)
```

```python
clusters_new.labels_
```

```python
#Assign clusters to the data set
Univ['clusters_new'] = clusters_new.labels_
```

```python
Univ
```

```python
#these are standardized values
clusters_new.cluster_centers_
```

```python
Univ.groupby('clusters_new').agg(['mean']).reset_index()
```

#DBSCAN

Density-Based Spatial Clustering of Applications with Noise

###Drawbacks of K-means and Hierarchical Clustering.
*   K-means algorithm requires one to specify the number of clusters a priory (the K value).
*   K-means does not perform well on finding non-convex/non-spherical shapes of clusters.
*   K-means is sensitive to noise data

##DBSCAN Parameters

### Epsilon or eps
Defines the size and borders of each neighbourhood. The e(must be higher than 0) is a radius. The neighbourhood of point x called the e-neighbourhood of x, is the circle/ball with radius e around point x.

###MinPts

The density threshold. If a neighborhood includes at least MinPts points, it will bne considered as a dense region. Alternatively, a point will be considered as dense if there are at least the value of MinPts points in its e-neighborhood. These dense points are called core points.

**Core Point** - A point will be called as core point if the size of its e-neighborhood is greater than equal to MinPts.


**Border Point** - A border point has e-neighborhood that contains less than MinPts points(so its not a core point), but it belongs to the e-neighborhood of another core point.



**Noise Point** - If a point isn't a core point and isn't a border point, its a noise point or an outlier.

##Directly Density Reachable

In order to answer, how does DBSCAN know which point goes to which cluster, we need to define some definitions.
Point y is directly density reachable from point x if :
*   Point y belongs to the e-neighborhood of point x.
*   and point x is a core point

##Density Reachable

Point y is density reachable from point x, if there is a path of points between point x and point y, where each point in the path is directly density reachable from the previous point. This means that all the points on the path are core points, with the possible exception of point y.

##Density Connected

A ponint x is density-connected to a point y, if there is a point O that both x and y are density-reachable from. This way the DBSCAN connects core objects and their neighborhood in a dense region.

##How to choose Best value for parameters

**eps**

In general, e should be chosed as small as possible.
Higher the e value, there will be less outliers.
The desirable amount of noise(outliers) will usually be between 1% and 30%.


**MinPts**

MinPts >= D+1, where is the number of dimensions in the dataset. The minimum value of MinPts must be chosen at least 3.

##DBSCAN PROS AND CONS

**PROS**
*   Indetifies randomly shaped clusters
*   doesn't necessitate to know the number of clusters in the data previously(as opposed to k-means)
*   Handles Noise

**CONS**
*   Datasets with varying densities are problematic
*   Input parameters (e and MinPts) may be difficult to determine
*   Computational complexity - when the dimensionality is high, it takes O(n²)

#DBSCAN Code

```python
#import the lib.
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Wholesale customers data.csv")
df.head()
```

```python
df.info()
```

```python
df.drop(['Channel','Region'],axis=1,inplace=True)
```

```python
array=df.values
```

```python
array
```

```python
stscaler = StandardScaler().fit(array)
X = stscaler.transform(array)
```

```python
X
```

```python
from sklearn.neighbors import NearestNeighbors
nn = NearestNeighbors(n_neighbors=8)
nbrs = nn.fit(X)
```

```python
dist, indices = nbrs.kneighbors(X)
```

```python
dist
```

```python
dist = np.sort(dist, axis=0)
```

```python
dist = dist[:,1]
```

```python
plt.plot(dist)
```

```python
dbscan = DBSCAN(eps=1,min_samples=3)
dbscan.fit(X)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['Clusters'])
```

```python
df['Cluster'] = cl
df
```

```python
df.drop(['Cluster'],axis=1,inplace=True)
df
```

```python
pd.concat([df,cl],axis=1)
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(df,df['Cluster'])
```
